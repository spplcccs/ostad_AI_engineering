### মডিউল-৩: ফাংশন পরিচিতি (Module-3: Introduction to Function)

**Function** (ফাংশন): পাইথনে কোড লেখার ক্ষেত্রে বা যেকোনো প্রোগ্রামিং ল্যাঙ্গুয়েজে ফাংশন একটি অত্যন্ত গুরুত্বপূর্ণ এবং শক্তিশালী কনসেপ্ট।

### ১. ফাংশন কী? (What is a Function?)

সহজ কথায় বলতে গেলে, ফাংশন হলো একটি নির্দিষ্ট কোড ব্লক (Block of code) যা তখনই রান করে যখন তাকে কল করা হয়।

* ধরুন, আপনার কোডে এমন কিছু কাজ আছে যা বারবার করা লাগে। আপনি যদি প্রতিবার সেই একই কোড বারবার লেখেন, তাহলে আপনার কোড অনেক বড় এবং অগোছালো হয়ে যাবে।
* এই সমস্যা সমাধানের জন্যই আমরা ফাংশন ব্যবহার করি। আমরা নির্দিষ্ট কাজের কোডটুকুকে একটা ব্লকের ভেতর রেখে তার একটি নাম দিয়ে দিই। পরবর্তীতে যখনই আমাদের ওই কাজটির প্রয়োজন হয়, আমরা শুধু সেই নামটি ধরে ডাক দিই বা কল করি।

#### ২. কেন আমরা ফাংশন ব্যবহার করবো? (Why use Functions?)

* **কোডের পুনব্যবহারযোগ্যতা (Code Reusability):** একবার কোড লিখে রাখলে তা বারবার যেকোনো জায়গায় ব্যবহার করা যায়। ফলে একই কোড বারবার টাইপ করতে হয় না।
* **সহজ রক্ষণাবেক্ষণ (Easy Maintenance):** কোডে কোনো পরিবর্তন বা বাগ (Bug) ফিক্স করতে হলে শুধু ফাংশনের ভেতরের কোডটুকু পরিবর্তন করলেই পুরো প্রজেক্টে তা আপডেট হয়ে যায়।
* **কোড ক্লিন রাখা (Clean and Organized Code):** ফাংশন ব্যবহারের ফলে কোড দেখতে অনেক পরিষ্কার, গোছানো এবং রিডেবল (Readable) হয়।

#### ৩. পাইথনে ফাংশন লেখার নিয়ম (Syntax of a Function)

পাইথনে একটি ফাংশন তৈরি করতে বা ডিফাইন করতে হলে আমাদের একটি কিওয়ার্ড ব্যবহার করতে হয়, সেটি হলো `def` (Definition-এর সংক্ষিপ্ত রূপ)।

এর সাধারণ গঠনটি নিচে দেওয়া হলো:

```python
def function_name():
    # ফাংশনের ভেতরের কোড ব্লক (Code block)
    print("Hello from a function!")
```

* প্রথমে `def` কিওয়ার্ড বসবে।
* এরপর আপনার পছন্দমতো ফাংশনের একটি নাম দিতে হবে (যেমন: `function_name`)।
* নামের পাশে একটি ফার্স্ট ব্র্যাকেট বা প্যারেন্থেসিস `()` দিতে হবে এবং শেষে একটি কোলন `:` দিতে হবে।
* কোলন দেওয়ার পর নিচের লাইনে একটু স্পেস বা ইনডেন্টেশন (Indentation) বজায় রেখে ফাংশনের ভেতরের মূল কাজ বা কোডটুকু লিখতে হবে।

#### ৪. ফাংশনকে কল করা (Calling a Function)

ফাংশন ডিফাইন বা তৈরি করলেই কিন্তু সেটি নিজে থেকে কোনো আউটপুট দেবে না। তাকে দিয়ে কাজ করাতে হলে তাকে "কল" করতে হবে। ফাংশন কল করার নিয়ম খুবই সহজ—শুধু ফাংশনের নামটি লিখে পাশে ব্র্যাকেট দিয়ে দিতে হয়:

```python
# ফাংশন কল করা
function_name()
```

### মডিউল-৩: বিল্ট-ইন ফাংশন এবং রিটার্ন বনাম প্রিন্ট (Module-3: Built-in Functions and Return vs Print)

আজকে আমরা আলোচনা করবো পাইথনের কিছু গুরুত্বপূর্ণ **Built-in Functions** এবং ফাংশনের অন্যতম দুটি গুরুত্বপূর্ণ স্টেটমেন্ট **Return** এবং **Print** এর মধ্যকার পার্থক্য নিয়ে।

### ১. বিল্ট-ইন ফাংশন কী? (What are Built-in Functions?)

পাইথনে কিছু ফাংশন আগে থেকেই তৈরি করা থাকে, যা ব্যবহার করার জন্য আমাদের আলাদা কোনো কোড লিখতে হয় না। এগুলোকে বিল্ট-ইন ফাংশন বলা হয়। যেমন:

* `print()`: কোনো আউটপুট স্ক্রিনে দেখানোর জন্য।
* `len()`: কোনো স্ট্রিং, লিস্ট বা ডেটার দৈর্ঘ্য (Length) বের করার জন্য।
* `type()`: কোনো ভ্যারিয়েবলের ডেটা টাইপ জানার জন্য।
* `input()`: ইউজারের কাছ থেকে ইনপুট নেওয়ার জন্য।

### ২. প্রিন্ট বনাম রিটার্ন (Print vs Return)

অনেকেই মনে করেন `print` এবং `return` একই কাজ করে, কারণ দুটোই কোনো না কোনোভাবে আউটপুট বা রেজাল্ট দেখায়। কিন্তু এদের কাজের মধ্যে বিশাল একটি পার্থক্য রয়েছে:

#### ক. প্রিন্ট (Print Statement)

* **কাজ:** `print()` এর কাজ হলো শুধুমাত্র কোনো তথ্য বা ফলাফলকে কনসোল বা স্ক্রিনে প্রদর্শন করা (To display something to the human user)।
* এটি কোনো ভ্যালু বা ডেটা কম্পিউটারের মেমোরিতে জমা রাখে না বা পরবর্তী কাজের জন্য কোনো মান ফেরত পাঠায় না।
* সহজ কথায়, এটি শুধু দেখার জন্য, একে নিয়ে পরবর্তীতে আর কোনো গাণিতিক বা লজিক্যাল কাজ করা যায় না।

#### খ. রিটার্ন (Return Statement)

* **কাজ:** `return` এর কাজ হলো একটি ফাংশনের এক্সিকিউশন শেষ করে তার ফলাফল বা ভ্যালুটিকে সেই জায়গায় ফেরত পাঠানো যেখানে ফাংশনটিকে কল করা হয়েছিল (To send a value back to the caller)।
* এটি স্ক্রিনে সরাসরি কিছু দেখায় না, বরং এটি ব্যাক-এন্ডে বা কম্পিউটারের মেমোরিতে একটি নির্দিষ্ট মান বা ডেটা পাঠিয়ে দেয়।
* এই রিটার্ন করা ভ্যালুটিকে আমরা নতুন একটি ভ্যারিয়েবলের মধ্যে সেভ (Save) করে রাখতে পারি এবং পরবর্তীতে অন্য কোনো হিসাব-নিকাশে বা অন্য কোনো ফাংশনের ইনপুট হিসেবে ব্যবহার করতে পারি।

### একটি বাস্তব উদাহরণ (Real-Life Analogy)

* **Print-এর উদাহরণ:** ধরুন আপনি একটা রেস্টুরেন্টে গিয়েছেন এবং শেফ (Chef) খাবার রান্না করে আপনার টেবিলের সামনে এনে শুধু আপনাকে দেখালো। আপনি শুধু চোখ দিয়ে দেখলেন, কিন্তু সেটা খেতে বা সাথে করে নিয়ে যেতে পারলেন না। এটাই হচ্ছে `print`।
* **Return-এর উদাহরণ:** শেফ খাবারটি রান্না করে একটি পার্সেল বক্সে ভরে আপনার হাতে দিয়ে দিল। এখন আপনি চাইলে সেটা তখনই খেতে পারেন, ফ্রিজে রেখে দিতে পারেন, অথবা অন্য কাউকে উপহার হিসেবেও দিতে পারেন। অর্থাৎ, খাবারটি এখন আপনার নিয়ন্ত্রণে। এটাই হচ্ছে `return`।

### মডিউল-৩: ইউজার ডিফাইন্ড ফাংশন (Module-3: User Defined Functions)

#### ১. ইউজার ডিফাইন্ড ফাংশন কী? (What is a User Defined Function?)

যেসব ফাংশন প্রোগ্রামার বা ডেভলপাররা তাদের নিজস্ব লজিক এবং কাজের সুবিধার্থে নিজেরা কোড লিখে তৈরি করে, সেগুলোকে ইউজার ডিফাইন্ড ফাংশন বলা হয়।

#### ২. প্যারামিটার এবং আর্গুমেন্ট (Parameters and Arguments)

একটি ফাংশনকে আরও ডাইনামিক বা কার্যকরী করার জন্য আমাদের ফাংশনের ভেতর ডাটা বা ভ্যালু পাঠাতে হয়। এখানে দুটি গুরুত্বপূর্ণ শব্দ আমাদের বুঝতে হবে:

* **প্যারামিটার (Parameter):** ফাংশন তৈরি বা ডিফাইন করার সময় ব্র্যাকেটের ভেতর যে ভ্যারিয়েবলগুলোর নাম আমরা লিখি, সেগুলোকে প্যারামিটার বলে। এটি ফাংশনের ভেতরে **লোকাল ভ্যারিয়েবল** হিসেবে কাজ করে।
* **আর্গুমেন্ট (Argument):** ফাংশনটিকে যখন আমরা কল করি, তখন ব্র্যাকেটের ভেতর যে আসল মান বা ভ্যালুগুলো (যেমন: কোনো সংখ্যা বা স্ট্রিং) পাস করি, সেগুলোকে আর্গুমেন্ট বলা হয়।

```python
#একটি সহজ উদাহরণ:

# এখানে 'name' হলো একটি Parameter
def greet_user(name):
    print("Hello, " + name + "! Welcome back.")

# এখানে "Sabin" হলো একটি Argument
greet_user("Sabin")
```

#### ৩. একাধিক প্যারামিটার ব্যবহার করা (Using Multiple Parameters)

আমরা চাইলে একটি ফাংশনে একের অধিক প্যারামিটার ব্যবহার করতে পারি। সেক্ষেত্রে কমা (`,`) ব্যবহার করে সেগুলোকে আলাদা করতে হয়। যেমন দুটি সংখ্যার যোগফল নির্ণয় করার জন্য একটি ফাংশন:

```python
def add_numbers(num1, num2):
    result = num1 + num2
    return result

# ফাংশন কল করে ভ্যালু প্রিন্ট করা
sum_result = add_numbers(10, 20)
print("The sum is:", sum_result)
```

এখানে `num1` এবং `num2` হলো প্যারামিটার, আর `10` এবং `20` হলো আর্গুমেন্ট। ফাংশনটি এই মান দুটিকে যোগ করে `return` স্টেটমেন্টের মাধ্যমে ফলাফলটি ফেরত পাঠাচ্ছে।

### মডিউল-৩: একাধিক আর্গুমেন্ট (Module-3: Multiple Arguments)

আজকে আমরা শিখবো কীভাবে একটি ফাংশনে **Multiple Arguments** বা একাধিক আর্গুমেন্ট নিয়ে কাজ করতে হয় এবং পাইথনে আর্গুমেন্ট পাস করার বিভিন্ন পজিশনাল বা কিওয়ার্ড টেকনিকগুলো কী কী।

বাস্তব জীবনের প্রজেক্টে অনেক সময় আমাদের এমন ফাংশন তৈরি করতে হয় যেখানে কতগুলো ইনপুট বা আর্গুমেন্ট আসবে তা আগে থেকে নির্দিষ্ট থাকে না, অথবা অনেকগুলো ডাটা একসাথে প্রসেস করতে হয়। চলুন এই বিষয়গুলো সহজভাবে বুঝে নিই:

#### ১. পজিশনাল আর্গুমেন্ট (Positional Arguments)

সাধারণত আমরা ফাংশন ডিক্লেয়ার করার সময় যেভাবে প্যারামিটার লিখি, কল করার সময় ঠিক সেই সিরিয়াল বা পজিশন অনুযায়ী আর্গুমেন্ট পাস করতে হয়। একে পজিশনাল আর্গুমেন্ট বলে।

* **উদাহরণ:** আপনার ফাংশনে যদি প্রথম প্যারামিটার `first_name` এবং দ্বিতীয় প্যারামিটার `last_name` হয়, তবে কল করার সময়ও প্রথমে ফার্স্ট নেম এবং পরে লাস্ট নেমই দিতে হবে। উল্টোপাল্টা করলে আউটপুটও উল্টে যাবে।

#### ২. কিওয়ার্ড আর্গুমেন্ট (Keyword Arguments)

পজিশনাল আর্গুমেন্টের সমস্যা দূর করার জন্য পাইথনে কিওয়ার্ড আর্গুমেন্ট ব্যবহার করা হয়। এখানে ফাংশন কল করার সময় প্যারামিটারের নাম উল্লেখ করে সরাসরি ভ্যালু অ্যাসাইন করে দেওয়া হয় (`parameter = value`)।

* এর সুবিধা হলো, আপনি যদি সিরিয়াল বা পজিশন মেইনটেইন নাও করেন, তবুও পাইথন সঠিক প্যারামিটারে সঠিক মানটিই পাঠাবে।
* **উদাহরণ:** `greet_user(last_name="Rahman", first_name="Sabin")`। এখানে পজিশন ঠিক না থাকলেও নাম উল্লেখ থাকায় কোড সঠিকভাবে রান করবে।



```python
#greet_user(last_name="Rahman", first_name="Sabin")
```

#### ৩. ডিফল্ট আর্গুমেন্ট (Default Arguments)

অনেক সময় আমরা চাই কোনো আর্গুমেন্ট যদি পাস করা নাও হয়, তাও ফাংশনটি যেন একটি নির্দিষ্ট মান ধরে নিয়ে কাজ সম্পন্ন করে। এর জন্য ফাংশন ডিফাইন করার সময়ই প্যারামিটারের একটি ডিফল্ট মান সেট করে দেওয়া যায়।

```python
def welcome_msg(name, country="Bangladesh"):
    print("Hello " + name + " from " + country)

# এখানে শুধু নাম পাস করলেও দেশ হিসেবে 'Bangladesh' নিয়ে নেবে
welcome_msg("Arif") 
```

#### ৪. আরবিট্রারি আর্গুমেন্ট বা `*args` (Arbitrary Arguments)

যদি আপনি আগে থেকে না জানেন যে ইউজার ফাংশনে কয়টি আর্গুমেন্ট পাস করবে, তখন প্যারামিটারের নামের আগে একটি স্টার বা অ্যাস্টেরিস্ক (`*`) চিহ্ন দিতে হয়। একে পাইথনে সাধারণত `*args` বলা হয়। এটি সমস্ত ইনপুটকে একটি **Tuple** হিসেবে রিসিভ করে।

### মডিউল-৩: কিওয়ার্ড আর্গুমেন্ট (Module-3: Keyword Arguments)

আসসালামু আলাইকুম সবাইকে। আশা করি সবাই ভালো আছেন।

আজকে আমরা বিশেষভাবে এবং গভীরভাবে আলোচনা করবো **Keyword Arguments** নিয়ে। পাইথনে কোড লেখার সময় কেন কিওয়ার্ড আর্গুমেন্ট এত জনপ্রিয় এবং এটি কীভাবে আমাদের কোডের রিডাবিলিটি (Readability) বহুগুণ বাড়িয়ে দেয়, তা আজ আমরা বিস্তারিত দেখবো।

চলুন কিওয়ার্ড আর্গুমেন্টের মূল বিষয়গুলো এবং এর প্রায়োগিক সুবিধাগুলো জেনে নিই:

#### ১. কিওয়ার্ড আর্গুমেন্ট কী? (What is a Keyword Argument?)

সহজ কথায়, কিওয়ার্ড আর্গুমেন্ট হলো এমন এক ধরণের আর্গুমেন্ট যা ফাংশন কল করার সময় প্যারামিটারের নাম উল্লেখ করে সরাসরি ভ্যালু অ্যাসাইন করে পাস করা হয়। অর্থাৎ, এখানে আমরা `parameter_name = value` এই ফরম্যাটটি ব্যবহার করি।

#### ২. পজিশনাল আর্গুমেন্টের সাথে পার্থক্য (Difference with Positional Arguments)

* **পজিশনাল আর্গুমেন্ট:** আপনি যেভাবে ফাংশন ডিফাইন করেছেন, ঠিক সেই সিরিয়াল বা পজিশন অনুযায়ী মান পাঠাতে হবে। যদি একটি ফাংশনে প্রথমে `name` এবং পরে `age` থাকে, তবে আপনাকে প্রথমে নাম এবং পরেই বয়স দিতে হবে।
* **কিওয়ার্ড আর্গুমেন্ট:** এখানে পজিশন বা সিরিয়াল কোনো ব্যাপারই না। আপনি চাইলে আগে `age` এবং পরে `name` উল্লেখ করে দিতে পারেন। যেহেতু আপনি নাম ধরে ভ্যালু দিচ্ছেন, পাইথন নিজে থেকেই বুঝে নেবে কোন মানটি কোন প্যারামিটারে যাবে।
  ধরুন আমাদের একটি ফাংশন আছে যা কোনো ইউজারের প্রোফাইল ইনফরমেশন দেখায়:

```python
def user_profile(username, role, status):
    print(f"User: {username} | Role: {role} | Status: {status}")

# কিওয়ার্ড আর্গুমেন্ট ব্যবহার করে কল করা (পজিশন পরিবর্তন করে)
user_profile(status="Active", username="Sabin", role="Admin")
```

এখানে লক্ষ্য করুন, ফাংশনের ডেফিনিশনে `status` ছিল সবার শেষে, কিন্তু কল করার সময় আমরা তা সবার প্রথমে দিয়েছি। কিন্তু নামের কিওয়ার্ড ম্যাচ করার কারণে পাইথন কোনো ভুল ছাড়াই সঠিকভাবে আউটপুট দেখাবে: `User: Sabin | Role: Admin | Status: Active`।

#### ৪. কিওয়ার্ড আর্গুমেন্টের প্রধান সুবিধাসমূহ (Key Benefits)

* **কোড বুঝতে সুবিধা হওয়া (Self-Documenting Code):** ফাংশনটি কল করার লাইন দেখেই যে কেউ সহজে বুঝতে পারে কোন ভ্যালুটা আসলে কীসের জন্য পাঠানো হচ্ছে।
* **ভুল এড়ানো (Avoiding Bugs):** ডেটার পজিশন উল্টোপাল্টা হয়ে গিয়ে আউটপুট বা লজিকে বড় কোনো ভুল হওয়ার সম্ভাবনা একদম থাকে না।
* **ডিফল্ট প্যারামিটারের সাথে চমৎকার কম্বিনেশন:** যখন কোনো ফাংশনে অনেকগুলো ডিফল্ট প্যারামিটার থাকে, তখন নির্দিষ্ট যেকোনো একটির মান পরিবর্তন করতে কিওয়ার্ড আর্গুমেন্ট সবচেয়ে বেশি সাহায্য করে।

বাস্তব ক্ষেত্রে যখন আমরা বড় প্রজেক্টে বা বিভিন্ন ফ্রেমওয়ার্কে (যেমন: Django বা FastAPI) কাজ করি, তখন কিওয়ার্ড আর্গুমেন্ট প্রতিনিয়ত ব্যবহার করতে হয়।

### মডিউল-৩: ডিফল্ট প্যারামিটার এবং পাস কিওয়ার্ড (Module-3: Default Parameter and Pass Keyword)

আজকে আমরা ফাংশনের আরও দুটি চমৎকার এবং বহুল ব্যবহৃত টপিক নিয়ে কথা বলবো। প্রথমটি হচ্ছে **Default Parameter** (ডিফল্ট প্যারামিটার) এবং দ্বিতীয়টি হচ্ছে পাইথনের একটি স্পেশাল কিওয়ার্ড, যার নাম **pass** (পাস কিওয়ার্ড)।

কোড লেখার সময় কোডকে ছোট করা এবং ফিউচার কোডের স্ট্রাকচার বা লেআউট তৈরি করার জন্য এই দুটি জিনিস জানা খুবই জরুরী। চলুন এক এক করে আলোচনা করা যাক:

#### ১. ডিফল্ট প্যারামিটার কী? (What is a Default Parameter?)

ফাংশন তৈরি বা ডিফাইন করার সময় আমরা যদি কোনো প্যারামিটারের একটি নির্দিষ্ট মান বা ভ্যালু আগে থেকেই অ্যাসাইন করে দিই, তবে তাকে ডিফল্ট প্যারামিটার বলা হয়।

* **সুবিধা:** ফাংশনটি কল করার সময় ইউজার যদি ওই নির্দিষ্ট প্যারামিটারের জন্য কোনো আর্গুমেন্ট বা ভ্যালু পাস না-ও করে, তবুও পাইথন কোনো error দেবে না। সে নিজে থেকেই ওই ডিফল্ট মানটিকে ধরে নিয়ে ভেতরের কাজ সম্পন্ন করবে।
* তবে ইউজার যদি নতুন কোনো মান পাস করে, তবে ডিফল্ট মানটি রিপ্লেস হয়ে ইউজারের দেওয়া নতুন মানটি সেট হয়ে যায়।

**কোড উদাহরণ:**

```python
def info(name, country="Bangladesh"):
    print(f"Name: {name}, Country: {country}")

# ১. দেশ পাস না করলে ডিফল্ট মান নেবে
info("Sabin") # আউটপুট: Name: Sabin, Country: Bangladesh

# ২. দেশ পাস করলে নতুন মান সেট হবে
info("John", "USA") # আউটপুট: Name: John, Country: USA
```

#### ২. পাস কিওয়ার্ড কী? (What is the 'pass' Keyword?)

পাইথনে লুপ, কন্ডিশনাল স্টেটমেন্ট কিংবা ফাংশনের বডি বা কোড ব্লক ফাঁকা (Empty) রাখা যায় না। আপনি যদি একটি ফাংশন ডিক্লেয়ার করে তার ভেতর কোনো কোড না লিখে ফাঁকা রেখে দেন, তবে পাইথন **IndentationError** বা syntax error দেখাবে।

* কিন্তু অনেক সময় এমন হয় যে আমরা একটি ফাংশন কেবল তৈরি করে রাখতে চাই, যার মূল লজিক বা কোড আমরা পরবর্তীতে লিখবো।
* এই ধরণের পরিস্থিতিতে ফাঁকা ফাংশনকে ইরর-মুক্ত রাখতে আমরা ফাংশনের ভেতরে **`pass`** কিওয়ার্ডটি ব্যবহার করি।
* `pass` শব্দের অর্থ হলো "কিছু না করে সামনে এগিয়ে যাও" (Do nothing / Null operation)। এটি কেবল একটি প্লেসহোল্ডার (Placeholder) হিসেবে কাজ করে।

```python
#def my_future_function():
```

```python
def my_future_function():
    pass # এটি দেওয়ার কারণে কোডে কোনো ইরর আসবে না
```

বাস্তব জীবনে যখন আমরা বড় টিম প্রজেক্টে কাজ করি, তখন আর্কিটেকচার বা স্ট্রাকচারাল লেআউট বানানোর জন্য `pass` কিওয়ার্ড এবং মেথড ওভারলোডিং এড়াতে `Default Parameter` আমাদের দারুণভাবে সাহায্য করে।

আসসালামু আলাইকুম সবাইকে। কী অবস্থা সবার? মডিউল ২ এর পর এবার আমরা মডিউল ৩ শুরু করতে যাচ্ছি, যেখানে আমরা শিখবো Function, File Handling এবং Error Handling। এখানে প্রতিটি টপিক বিস্তারিতভাবে আলোচনা করা হবে যাতে আপনার কোডিং দক্ষতা আরও উন্নত হয়।

চলুন দেখে নিই মডিউল ৩ এর ক্লাস এবং সময়:

* **প্রথম ক্লাস:** শুক্রবার দুপুর ৩:৩০ মিনিটে। ইনস্ট্রাক্টর হিসেবে থাকবেন তামিম শাহরিয়ার সুবিন ভাইয়া।
* **দ্বিতীয় ক্লাস:** রবিবার রাত ৯:০০ টায়। ইনস্ট্রাক্টর হিসেবে থাকবেন আব্দুল্লাহ আল নাইম ভাইয়া।

#### আমরা কী কী শিখবো? (What We Will Learn)

**১. ফাংশন (Function)**

* কীভাবে পাইথনে ফাংশন লিখতে হয়।
* ফাংশন কী এবং কীভাবে ফাংশনের কোড ব্লককে পুনরায় ব্যবহার (Reuse) করা যায়।
* ফাংশনের প্যারামিটার (Parameter) কীভাবে লিখতে হয়।
* ফাংশনের মধ্যে আর্গুমেন্ট (Argument) পাস করা, তাদের ব্যবহার এবং ভ্যালু পাস করার কৌশল।
* এছাড়া Keyword Argument — যেখানে আর্গুমেন্ট পাস করা এবং এটি কীভাবে কোডকে আরও ক্লিন (Clean) করে তোলে, সে বিষয়গুলো জানবো।
* Return Statement সম্পর্কে জানবো।
* সংক্ষিপ্ত ফাংশন লিখতে Lambda Expression ব্যবহার করা হয়, সেটি কীভাবে করা হয় সেটিও শিখবো আমরা।
* Local and Global Scope বা ভ্যারিয়েবলের লাইফ-সাইকেল নিয়ে আলোচনা করা হবে।

**২. ফাইল হ্যান্ডলিং (File Handling)**

* কীভাবে পাইথন কোড ব্যবহার করে কম্পিউটারের যেকোনো ফাইল (যেমন: `.txt` বা `.csv`) ওপেন করা, পড়া (Read), নতুন তথ্য লেখা (Write) বা ফাইল বন্ধ (Close) করা যায়।

**৩. এরর হ্যান্ড信ং বা ব্যতিক্রম হ্যান্ডলিং (Error & Exception Handling)**

* কোড রান করার সময় কোনো ভুল বা ক্র্যাশ হলে তা কীভাবে `try-except` ব্লকের মাধ্যমে চমৎকারভাবে হ্যান্ডেল করা যায়, যেন পুরো সফটওয়্যারটি বন্ধ না হয়ে যায়।
